## Broadcasting in arrays
- array rules
- applying one value to each and every element

```python
l1=[10,20,30]
a1=np.array(l1)
print(a1)
a1.shape
```

```python
c=100
a1+c
```

```python
l2=[11,12,13]
a2=np.array(l2)
print(a2)
a2.shape
```

```python
a1+a2
```

```python
a3=np.array([100,200,300,400])
# a1+a3 ValueError: operands could not be broadcast together with shapes (3,) (4,)
```

***Rules for broadcasting*** 
- elements must be equal(shape) eg. a1.shape=(3,2) and a2.shape(3,2), always compares from right to left.
- Checks for the rules (2==2, 2==1, 2==empty)
- If all the conditions are true then only it performs operations

```python
a4=np.array([[1,2,3],[5,6,7]])
print(a4)
print(a4.shape)
```

```python
a1+a4
```

### linspace( )
- It is going to divide the numbers into differnent equal parts
- By default it will use 50 parts to divide equally
- It includes the stop value also

```python
np.linspace(0,10,5,retstep=True) # retstep returns the step value 
np.linspace(0,10,5,retstep=True,endpoint=False) # if i dont want to include the endvalue then use endpoint=False
```

```python
rd.randint(2)
```

```python
rd.randint(10,100,(3))
```

- If i run the program again and again it will be generating new numbers everytime to fix to same values we use seed( )

```python
rd.seed(2)
rd.randint(10,100,(3,2))
```

- rand( ) method generates only positive floating numbers between 0 and 1

```python
# rd.seed(2)
rd.rand(3)
```

```python
rd.rand(2,3)
```

- randn( ) will generate a random floating numbers between -1 - to - 1 which includes both postive as well as negative value  

```python
rd.randn(2)
```

```python
r=rd.randn(2)
print(r)

print(np.mean(r))
print(np.std(r))
```

```python
r=rd.randn(2,3)
print(r)

print(np.mean(r))
print(np.std(r))
```

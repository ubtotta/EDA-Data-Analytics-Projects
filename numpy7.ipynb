```python
import numpy as np
```

```python
a1=np.array([10,20,30,])
print(a1)
```

- power( ) function
- for each element in the array it will multiply the value itself, based on the power value given

```python
np.power(a1,2)
```

- sum( ) function
- sum( ) will add all the elements in the array 

```python
np.sum(a1)
```

- std( ) function gives the standard deviation of the array passed

```python
np.std(a1)
```

- mean( ) gives the average value of given array

```python
np.mean(a1)
```

- var( ) function 
- used to check the how far the value is from the mean value
- if variance value is less, near to the mean value
- if variance value is high, far from the mean value

```python
np.var(a1)
```

```python
a2=np.array([1,0,2])
print(a2)
```

- exp( ) function
- gives the power value

```python
np.exp(a2)
```

- log( ) functions
- gives the value of that power

```python
a3=np.array([2.71828183, 1.        , 7.3890561 ])
np.log(a3)
```

- isin ( ) - it works like a membership operator to check wheather the element is present in the array or not

```python
arr1=np.array([10,2,15,5,9,20])
print(arr1)
```

```python
20 in arr1
```

```python
np.isin(20,arr1)
```

```python
np.isin([20,5,15,666],arr1)
```

- sort( ) - bydefault arranges in ascending order

```python
np.sort(arr1)
```

- median( ) - return the mid value

```python
np.median(arr1)
```

```python
import numpy as np
```

```python
a1=np.array([10,20,30])
print(a1)
```

## copy( )
- copy is independent

```python
s_a=np.copy(a1)
print(s_a)

s_a[0]=100
print(s_a)
print(a1)
```

## view( )

- view( ) will also copy the elements, but it points to same memory location.
- view is dependent

```python
a3=a1.view()
print('a3:' ,a3)
print('a1:' ,a1)

a3[-1]=200
print('a3:', a3)
print('a1:',a1)
```

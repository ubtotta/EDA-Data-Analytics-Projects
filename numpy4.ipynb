```python
import numpy as np
```

```python
l1=[1,2,3,]
a1=np.array(l1)
a1
```

```python
l2=[6,7,8]
a2=np.array(l2)
a2
```

## Arithematic operations

```python
print(a1+a2)
# adds the corresponding elements inside the array.
```

```python
print(a1-a2)
print(a1*a2)
print(a1**a2)
print(a1/a2)
print(a1%a2)
print(a1//a2)
```

## Comparison operators

```python
print(a1==a2)
print(a1!=a2)
print(a1<a2)
print(a1<=a2)
print(a1>a2)
print(a1>=a2)
```

```python
print((a1>5).all() and (a1<10).any())
```

```python
import numpy as np
```

```python
a1=np.arange(10,96,5)
print(a1)
```

```python
print(a1.shape)
```

-  When we have to reshape an array we have to check with the multiples of the array values length. 
- For eg. 18 is size of the array , we have 18 elements, so multiples of 18 are 2X9, 3X6 etc.

```python
np.reshape(a1, (3,6))
```

```python
a2=np.reshape(a1,(3,2,3))
print(a2)
```

- when i pass the reshape size as -1 then it will convert any shape of array to the 1D array.

```python
np.reshape(a2,(-1))
```

```python
a3=np.array([[10,20,30,40,50,78],[2,4,5,60,64,28],[11,9,37,92,55,68]])
print(a3)
print(a3.shape)
print('Dimension: ',a3.ndim)
print('Size: ', a3.size)
```

```python
np.reshape(a3,(3,2,1,3))
np.reshape(a3, (3,3,2))
```

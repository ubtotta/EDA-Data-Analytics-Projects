## melt( ) function - it will convert columns into rows

- when we want to apply aggregate functions or group by clause or when we want to visualize or 

when we need it for machine learning process we need to convert it into to rows.

#### Syntax
pd.melt(DataFrame, id_vaiables, value_variables, variable_names, value_names)

- id_vars (id_variables) = Columns doesn't want to change.

- value_vars (value_variables) = Columns which we want to convert into the rows.

- var_name (vaiable_names) = name used to give to column which we are converting (column name for old columns).

- value_name (value_names) = name for the value columns 

```python
pd.melt(df,value_vars=['Hotstar','Netflix'])
```

```python
pd.melt(df,value_vars=['Hotstar','Netflix'],var_name='OTT',value_name='Available')
```

```python
df1=pd.melt(df,id_vars=['Movies','Rating'],value_vars=['Hotstar','Netflix'],var_name='OTT',value_name='Available')
```

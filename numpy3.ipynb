## Accessing list elements using index values

```python
import numpy as np
```

```python
lst1=[10,20,30,40,50]
```

```python
lst2=[[10,20,30,40],[50,60,70,80],[90,89,72,11]]
```

```python
lst1[1]
```

```python
lst2[1][1]
```

```python
# lst2[1,1]
# TypeError: list indices must be integers or slices, not tuple
```

```python
arr1=np.array(lst1)
arr1
```

```python
arr2=np.array(lst2)
arr2
```

## Slicing method in arrays
- arr[start : stop : step]

```python
lst1[::-1]
```

```python
lst2[::-1]
# it reverse only the outer elemens not the inner elements
```

```python
lst2[::-1]
```

```python
arr2[::, ::-1]
```

```python
arr2[::-1]
```

```python
arr2[::-1,2] #it will select the whole column
```

```python
arr2[::-1,::-1]
```

```python
arr2[::-1][2] # it will select the whole row
```

```python
arr2[::2]
```

```python
arr2[::2,::2]
```

```python
arr2[:1:,::2]
```

```python
lst3=[[[10,20],[30,40]],[[1,2],[50,60]]]
```

```python
arr3=np.array(lst3)
arr3
```

```python
arr3[1,0,1]
```

```python
arr3[::,::,::]
```

```python
arr3[::-1,:1:,::2]
```

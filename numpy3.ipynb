## Accessing list elements using index values

```python
import numpy as np
```

```python
lst1=[10,20,30,40,50]
```

```python
lst2=[[10,20,30,40],[50,60,70,80],[90,18,21,11]]
```

```python
lst1[1]
```

```python
lst2[1][1]
```

```python
# lst2[1,1]
# TypeError: list indices must be integers or slices, not tuple
```

```python
arr1=np.array(lst1)
arr1
```

```python
arr2=np.array(lst2)
arr2
```

- arr[ row , col ]
- arrays can accept the multi index.

```python
arr1[2]
```

```python
arr2[1,1]
```

```python
arr2[1][3]
```

## Slicing method in arrays
- arr[start : stop : step]

```python
lst1[::-1]
```

```python
lst2[::-1]
# it reverse only the outer elemens not the inner elements
```

```python
lst2[::-1]
```

```python
arr2[::, ::-1]
```

```python
arr2[::-1,::-1]
```

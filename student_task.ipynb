```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading the dataset

```python
df=pd.read_csv('student_performance_dirty.csv')
```

```python
df
```

## Checking the data set using info( )

```python
df.info()
```

## Now checking for any duplicate values

```python
df.duplicated().any()
```

```python
df.duplicated(subset='Age').any()
```

## Droping the duplicate records from the DataFrame

```python
df=df.drop_duplicates(subset='StudentID')
df
```

```python
df.duplicated().any()
```

## Now check for the null values

- We have some records missing in the Age column and marks column
- We will check for outliers using boxplot

```python
sns.boxplot(df)
plt.show()
```

- As we can see in above graph we have a one outlier in the marks category so we have fill the data with median 

```python
age_mean=df['Age'].mean()
marks_median=df['Marks'].median()
```

## Now filling the null values 

```python
df['Age']=df['Age'].fillna(age_mean)
df['Marks']=df['Marks'].fillna(marks_median)
```

```python
df
```

```python
df.info()
```

```python
df=df.astype({'Age':'int'})
```

```python
df.info()
```

## Analysis

1.What is the gender distribution of students? Are male and female students equally represented?
Hint:Count of Male and Female students (Countplot)

```python
plt.figure(figsize=(4,4))
c=sns.countplot(df,x='Gender',width=0.5)
plt.bar_label(c.containers[0])
plt.show()
```

- No male and females are not distributed equally, as we can see from above graph there are 6 males and 5 females.

2.What percentage of students passed and failed? Provide a quick summary for management.
Hint:Pass vs Fail percentage (Pie Chart)

```python
c=df['Pass'].value_counts()
print(c.values)
print(c.index)
```

```python
plt.figure(figsize=(4,4))
plt.pie(x=c.values, labels=c.index, autopct='%.1f%%', explode=[0,1])
plt.title('Pass Vs. Fail Percentage')
plt.legend()
plt.show()
```

3.Which grade category has the highest average marks? Compare the average marks across all grades.
hint:Average Marks by Grade (Bar Plot)

```python
c=sns.barplot(df, x='Grade', y='Marks', errorbar=None, width=0.6)
plt.bar_label(c.containers[0],fmt='%.1f%%' )
plt.yticks(range(0,112,10))
plt.title('Highest Avg. marks in Grades')
plt.show()
```

- A grade has the highest average marks

4.Does spending more time studying lead to better marks? Analyze the relationship between study hours and marks.
hint:Study Hours vs Marks (Scatter Plot)

```python
plt.figure(figsize=(5,5))
sns.scatterplot(df,x='StudyHours',y='Marks')
plt.title('Relation between Study hours and Marks')
plt.show()
```

- By looking at the above scatter plot we can conclude that with increase in the study hours there is increment in the marks 

5.Is there any difference in pass and fail rates between male and female students?
hint:Gender-wise Pass/Fail (FacetGrid)

```python
fc = sns.FacetGrid(data=df, col='Gender', aspect=1.1)
fc.map_dataframe(sns.countplot, x='Pass')
plt.show()
```

- From above graph we can observe in male category there is only one male who is failed and in females category all are passed.

6.How are students' marks distributed? Are most students scoring high, average, or low marks?
Hint:Distribution of Marks (Histogram)

```python
plt.figure(figsize=(5,5))
c=sns.histplot(data=df, x='Marks',kde=True, bins=6)
plt.bar_label(c.containers[0])
plt.title("Distribution of Students' Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

7.Are there any unusually high or low marks that could affect our analysis?
hint:Detect Outliers in Marks (Boxplot)

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Marks', color='lightgreen')
plt.title("Boxplot of Students' Marks")
plt.xlabel("Marks")
plt.show()
```

- Marks column has outliers.

8.Which gender has better average attendance? Compare attendance across genders.
hint:Average Attendance by Gender (Bar Plot)

```python
plt.figure(figsize=(5,5))
c=sns.barplot(df,x='Gender',y='Attendance', errorbar=None, width=0.5)
plt.bar_label(c.containers[0], fmt='%.1f%%')
plt.show()
```

- As per observation female has more average in attendance percentage.

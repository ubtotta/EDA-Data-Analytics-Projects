```python
d2={'name':['scott','martin','adams','martin','miller'],
    'sal':[3000,7000,1000,6000,5000],
    'comm':[None,None,100,None,500],
    'deptno':[10.0,30.0,30.0,20.0,30.0],
    'hiredate':['12/apr/2000','1/may/2000','10/aug/2000','12/dec/2000','12/jan/2002'],
    }

df2=pandas.DataFrame(d2)
```

## concat()

```python
#by default axis=0(joins the data row by row)
df=pandas.concat([df1,df2], axis=0)
df
```

```python
df=df.reset_index(drop=True) #to reset all the index and to drop the previous index use Drop=True reset_index()
df
```

```python
df3=pandas.DataFrame({'column_x':[0]*10,
                      'column_y':[None]*10})
df3
```

```python
df=pandas.concat([df,df3],axis=1)
```

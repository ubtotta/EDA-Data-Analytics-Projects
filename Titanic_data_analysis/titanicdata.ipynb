## Merge ( )
- Syntax
- pd.merge(tablename1, tablename2, on='common_column' ,left_on=' ', right_on=' ' ,how = ' ')
- if the column name are different and values are same from both the tables then we use left_on and right_on and in how we can pass, inner or left or right.

```python
df=pd.merge(df,ticket_pass,on='ticket',how='inner')
```

```python
df
```

```python
df['fare']=(df['fare']/df['NumOfPasgs']).round(1)
```

```python
df
```

```python
def comp_type(row):
    f=row['Fam_Count']
    p=row['NumOfPasgs']
    if p==1:
        return 'Solo'
    elif f==0 and p>1:
        return 'Friends'
    else:
        return 'Family'
```

- apply( ) is a grouping function, it iterates like a for loop in python based on the axis , if axis=1 applies row-by-row or if axis=0 applies.column-by-column.
- pandas will treat row like a key value pair
- when ever we have complex condition we use apply( ) function.

```python
df['Companion type']=df.apply(comp_type,axis=1)
df
```

```python
def age_category(row):
    age=row['age']
    if age>0 and age<=2:
        return 'Infant'
    elif age>2 and age<=16:
        return 'Child'
    elif age>25 and age<=59:
        return 'Adult'
    else:
        return 'Senior'
```

```python
df.apply(age_category,axis=1)
```

- cut( ) : when we have only numerical data's and continuos values then we can apply cut( ) function.
- Syntax
- pd.cut(column_name, bins=[ranges], labels=[returning value])

```python
df['Age_category']=pd.cut(df['age'],bins=[0,2,16,25,59,100],labels=['Infant', 'Child', 'Teen', 'Adult', 'Senior'])
df
```

```python
Royals=['the Countess', 'Jonkheer', 'Don', 'Dona', 'Master', 'Sir']
Common=['Miss', 'Mr', 'Mrs', 'Lady', 'Ms', 'Mme', 'Mlle']
Military=['Major', 'Col', 'Capt']
Other_professionals=['Dr', 'Rev']
```

```python
def social_status(row):
    c=row['Title']
    if c in Royals:
        return 'Royals'
    elif c in Common:
        return 'Common'
    elif c in Military:
        return 'Military'
    else:
        return 'Other Professionals'
```

```python
df['Social Status']=df.apply(social_status,axis=1)
df
```

## Plots
### 1) Univariant Plot - which plot with one variable (one column) 
   - CountPlot(Number), piechart(Percentage distribution)

3) Multi-variant - which accepts multiple variable (multiple columns)

```python
plt.figure(figsize=(3,4))
sns.countplot(data=df, x='gender', width=0.6, color='red')
plt.title('No. of passengers gender wise')
plt.yticks(range(0,1000,100)) #for adjusting the ranges
plt.ylabel('No of Passengers')
plt.show() #it will show only the graph, it will ignore all the other text in the output box.
```

**Observation**
- By looking at the graph i can conclude that there are more number of male passengers travelling compared to female passengers.

```python
plt.figure(figsize=(4,5))
cnt=sns.countplot(data=df, x='pclass', width=0.6, color='pink', order=[3,1,2])
cnt.bar_label(cnt.containers[0])
plt.title('No of passengers in each class')
plt.yticks(range(0,900,100))
plt.ylabel('No of passengers')
plt.show()
```

**Observation**
- By looking at the graph i can conclude that most of the passengers are from the class '3' and second most is from the class '1' and third is class '2'.


```python
plt.figure(figsize=(5,4))
cnt=sns.countplot(data=df, x='Age_category', hue='gender', palette=['cyan', 'crimson'])
for i in cnt.containers:
    cnt.bar_label(i)
# cnt.bar_label(cnt.containers[0])
plt.title('Distribution of psgs according to age category')
plt.yticks(range(0,1000,100))
plt.ylabel('No of passengers')
plt.show()
```

**Observation** 
- By the above graph i can conclude that adults are travelling more in numbers.

- autopct is used to show the percentage in the pie chart for each category.

### 2) Bivariant Plot - which plot with two variables (two columns)
     
     -Barplot- always used to find the aggregative values, we have to use it with categorical data and numerical data.
       In estimator we can pass the aggregator functions for the plot like mean, max, min etc.

```python
plt.figure(figsize=(3,4))
c=sns.barplot(data=df, x='gender', y='fare', width=0.5, errorbar=None, estimator='mean',color='crimson')
plt.title('Avg. fare by gender wise')
plt.yticks(range(0,20,2))
plt.bar_label(c.containers[0], fmt='%.1f')
plt.show()
```

- Observation: 
  - on an average females are paying more fare
  - female passengers has higher average fare , then compared to male.

```python
plt.figure(figsize=(3,4))
c=sns.barplot(data=df, x='gender', y='fare', width=0.5, errorbar=None, estimator='sum', color='cyan')
plt.title('Total fare by gender wise')
plt.yticks(range(0,13000,2000))
plt.bar_label(c.containers[0], fmt='%.1f')
plt.show()
```

- Observation:
  - Male passengers are paying higher fare compared to the females.

```python
# fare and age category
plt.figure(figsize=(6,4))
c=sns.barplot(data=df, x='Age_category', y='fare', width=0.7, errorbar=None, estimator='mean', palette=['crimson','silver'], hue='gender')
for i in c.containers:
    c.bar_label(i, fmt='%.1f')
plt.yticks(range(0,36,5))
plt.legend(loc='best', ncol=2, bbox_to_anchor=(0.49,1), title='Gender')
plt.title('Avg. fare of psgs. according to Age_category')
plt.show()
```

- Observation:
  - Avg. fare increase with the age, among all the categories females pays consistently higher average fare than compared to male.

### - Subplot
- dividing the page into multiple sections and ploting multiple graphs is called as subplot

```python
g=df.loc[df['Companion type']=='Family'] #only storing the records of family
gf=g['gender'].value_counts()
gn=g['Age_category'].value_counts()
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2, figsize=(7,4))
plt.suptitle('Companion type wise categorization')
axes[0].pie(x=gf.values, labels=gf.index, autopct='%.1f%%')
axes[0].set_title('Gender wise categorization')
axes[1].pie(x=gn.values, labels=gn.index, autopct='%.1f%%')
axes[1].set_title('Age Category wise categorization')
plt.tight_layout()
plt.show()
```

- Obervation  
  - In a family more passengers are females as female percentage is higher and most of the passengers are adult compared to other category

## Co-relation

- Visualizing numerically.
- finds the relation between two or more columns and how strongly they are corelated.
- we can apply the df.corr( ) only on numerical values.

## Heatmap

- If we want to visualize the corr( ) using graphs then use heatmap in seaborn module.
- When the colour is dark then it has strong relation.
- When the color is light then it has weak relation.

```python
c=df.corr(numeric_only=True)
c
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(c, annot=True,fmt='.2f')
```

```python
import numpy as np
```

```python
a1=np.array([[1,2,3],[10,20,30]])
print(a1)
```

```python
np.reshape(a1,(-1))
```

## ravel( )
- it also converts any shape of array into 1D array.
- it behaves like a view, it is dependent, when we change the value in updated array its also making the changes original array due to common memory location.

```python
x=a1.ravel()
# print(x)

x[0]=50
print(a1,x,sep='\n')
```

```python
a4=np.array([[11,22,33],[4,67,80]])
print(a4)
```

## flatten( )
- it also behaves like a reshape, it also converts any shape of array to 1A array
- it is independent

```python
y=a4.flatten()
print(y)

y[-1]=90
print(a4,y,sep='\n')
```
